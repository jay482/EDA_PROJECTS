## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Writing my code to make my dataset analysis ready.


#first im changing the  column that i need in my data set
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                          'region_txt':'Region','attacktype1_txt':'Attack_Type','target1':'Target','nkill':'Killed',
                          'nwound':'Wounded','gname':'Group','targtype1_txt':'Target_Type',
                          'weaptype1_txt':'Weapon_type','latitude':'Latitude',
                          'longitude':'Longitude','target1':'Target','city':'City'},inplace=True)
```

```python
#HERE IM KEEPING THE COLUMN THAT I NEED IN THIS DATA FRAME
new_df=df[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','Attack_Type','Killed',
              'Wounded','Group','Target','Target_Type','Weapon_type']]
new_df
```

```python
#finding the null values in new data set
new_df.isnull().sum()
```

```python
#filling the null values on all categorical columns data frame for clean data and performong on the all the column
a=new_df['State'].mode()[0]# [0] index is used to extract the mode value from the resulting Series, in case there are multiple modes thats why we use it.
new_df['State']=new_df['State'].fillna(a)
```

```python
#PERFORMING ON ALL COLUMNS
a=new_df['City'].mode()[0]# [0] index is used to extract the mode value from the resulting Series, in case there are multiple modes thats why we use it.
new_df['City']=new_df['City'].fillna(a)
```

```python
'''The method i used spaital interpolation
These methods estimate missing values based on the values of neighboring locations. Techniques like inverse distance weighting (IDW), kriging, or nearest neighbor
interpolation can be used to estimate missing values based on the known values of nearby locations.'''
#THIS IS MOST IMPORTANT METHOD FOR DEALING GEOGRAPHICAL When dealing with geographic data like longitude and latitude, spatial interpolation techniques can be employed
# Method 2: Spatial interpolation (Nearest neighbor)  'nearest' method to fill the null values based on the nearest available non-null value in the respective column.
new_df['Longitude'] = new_df['Longitude'].interpolate(method='nearest')
new_df['Latitude'] = new_df['Latitude'].interpolate(method='nearest')
```

```python
#PERFORMING ON ALL COLUMNS
a=new_df['Target'].mode()[0]# [0] index is used to extract the mode value from the resulting Series, in case there are multiple modes thats why we use it.
new_df['Target']=new_df['Target'].fillna(a)
```

```python
#Removing Unknown value from data set beacuae i dont know where it belongs
new_df = new_df.drop(new_df[new_df['Group'] == 'Unknown'].index)
```

```python
new_df['Killed'] = new_df['Killed'].fillna(0)
new_df['Wounded'] = new_df['Wounded'].fillna(0)
```
